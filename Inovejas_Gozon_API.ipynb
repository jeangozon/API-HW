```python
sources = newsapi.get_sources(country='ph')
sources
```

```python
all_news = newsapi.get_everything(q='Philippines',
                                    from_param='2022-05-24',
                                    to='2022-05-24',
                                    language ='en',
                                 exclude_domains='youtube.com,facebook.com',
                                 page_size=50,
                                 sort_by='relevancy')
all_news
```

```python
data = pd.DataFrame(all_news)
data
```

```python
data.info()
```

```python
data1 = data.drop(['status','totalResults'],axis = 1) 
# axis 1 = specifies that the status and results are columns ; 
# 1 = column; 0 = row
# other option: data1 = data.drop(columns=['status','totalResults'])
data1
```

```python
data1 = pd.concat([data1.drop(['articles'], axis=1), 
                  data1['articles'].apply(pd.Series)], axis=1)
data1
# code descriptions:
# concat = https://pandas.pydata.org/docs/reference/api/pandas.concat.html
# drop = https://www.w3resource.com/pandas/dataframe/dataframe-drop.php
# series = https://pandas.pydata.org/docs/reference/api/pandas.Series.html
# apply = https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html
```

```python
data1 = data1.rename(columns = {'publishedAt':'Publication Date',
                                'title':'Title',
                               'content':'Article Content',
                               'author':'Author',
                               'url':'Article URL'})
data1
```

```python
data1 = data1[['Publication Date','Title','Article Content','Author','Article URL']]
data1
# NOTE: ADDED URL COLUMN FOR BETTER DATA USE BY NEXT USER
# rearranging dataframe columns
```
